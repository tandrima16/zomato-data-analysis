```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('zomato.csv')
```

```python
data.head()
```

```python
data.sample(5)
```

```python
data.info()
```

```python
data.isna().sum() 
```

removing unnecessary columns like url, address and phone.
rest_type is similar to listed_in(type)
location is similar to listed_in(city)
removing location and rest_type

```python
del data['url']
del data['address']
del data['phone']
del data['rest_type']
del data['location']
del data['menu_item']
```

renaming the columns for convinience.

```python
data.rename(columns={"approx_cost(for two people)" : "approx_cost", "listed_in(type)" : "dining_type", 
                   "listed_in(city)" : "city"}, inplace=True)
```

```python
data.sample(5)
```

```python
data.isna().sum() # dish liked contains null values but we cannot drop them as we will lose 50% of our data
```

removing the null values

```python
data['approx_cost'] = data['approx_cost'].astype(str) 
data['approx_cost'] = data['approx_cost'].apply(lambda x: x.replace(',',''))
data['approx_cost'] = data['approx_cost'].astype(float) 
```

```python
data['approx_cost'] = data['approx_cost'].fillna(data['approx_cost'].mean())
```

```python
#filling all the missing data with unrated
```

```python
data['rate'] = data['rate'].apply(lambda x: str(x).split('/')[0])
```

```python
a = list(data['rate'])
for i in range(0, len(a)):
    if a[i] == 'nan':
        a[i] = 'unrated'
    elif a[i] == '-':
        a[i] = 'unrated'
    elif a[i] == 'NEW':
        a[i] = 'unrated'
```

```python
data['rate'] = a
```

```python
# replacing , from approx cost and converting it into float data type
```
